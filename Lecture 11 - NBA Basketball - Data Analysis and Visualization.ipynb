Iterating through each and every matrix in this lecture is a repetitive process. This can be simplified with a help for a function which can be reused according to the needs.

```python
#Function takes the Matrix to visualize and the PlayerList as inputs. By default, Salary matrix will be visualized and Players list is passed
def customplot(InputMatrix = Salary, PlayerList = Players):
    ColorList = { "KobeBryant":"Black","JoeJohnson":"Red","LeBronJames":"Green","CarmeloAnthony":"Blue","DwightHoward":"Magenta","ChrisBosh":"Black","ChrisPaul":"Red","KevinDurant":"Green","DerrickRose":"Blue","DwayneWade":"Magenta" }
    MarkerList = { "KobeBryant":"s","JoeJohnson":"o","LeBronJames":"^","CarmeloAnthony":"D","DwightHoward":"s","ChrisBosh":"o","ChrisPaul":"^","KevinDurant":"D","DerrickRose":"s","DwayneWade":"o" }
    for name in PlayerList:
        plt.plot(InputMatrix[Pdict[name]], c=ColorList[name], ls='--', marker=MarkerList[name], ms=5, label=name)
    plt.xticks(list(range(0, 10)), Seasons)
    plt.legend(loc = 'upper left', ncol = 2, bbox_to_anchor=(1,1))
    plt.show()
```

Explanation for the above custom function
- **customplot()** function takes a Matrix to visualize as InputMatrix and the list of players as PlayerList, as inputs
- **ColorList** internal list is created to assign a particular color to each of the player, so that the graph lines are consistent
- **MarkerList** internal list is created to assign a particular marker style to each of the player, so that the graph markers are consistent
- Plots are created by iterating through the PlayerList input and for the particular InputMatrix
- Other plot aesthetics are performed

```python
#let's just call the customplot() function which should visualize Salary for all the players by default
customplot()
```

```python
#same as the above function call
customplot(Salary, Players)
```

```python
#let's just call the customplot() function with a list of selected players which should visualize Salary by default
customplot(PlayerList = ["KobeBryant","JoeJohnson","LeBronJames"])
```

```python
#let's just call the customplot() function for Games matrix which should visualize for all the players by default
customplot(InputMatrix = Games)
```

```python
#let's just call the customplot() function for Games matrix for a list of selected players
customplot(Games, ["KobeBryant","JoeJohnson","LeBronJames"])
```

```python
#let's just call the customplot() function for FieldGoalsPerGame matrix for a list of selected players
customplot(FieldGoalsPerGame, ["KevinDurant","DerrickRose","DwayneWade"])
```
