```python
#let's visualize salary trend
for i in list(range(0, len(Players) - 1)):
    plt.plot(Salary[i], ls='--', marker='s', ms=5, label=Players[i])
plt.xticks(list(range(0, 10)), Seasons)
plt.legend(loc = 'upper left', ncol = 2, bbox_to_anchor=(1,1))
plt.show()
```

```python
#let's visualize games trend
for i in list(range(0, len(Players) - 1)):
    plt.plot(Games[i], ls='--', marker='s', ms=5, label=Players[i])
plt.xticks(list(range(0, 10)), Seasons)
plt.legend(loc = 'upper left', ncol = 2, bbox_to_anchor=(1,1))
plt.show()
```

```python
#let's visualize accurancy trend
for i in list(range(0, len(Players) - 1)):
    plt.plot(FieldGoalAccuracy[i], ls='--', marker='s', ms=5, label=Players[i])
plt.xticks(list(range(0, 10)), Seasons)
plt.legend(loc = 'upper left', ncol = 2, bbox_to_anchor=(1,1))
plt.show()
```
