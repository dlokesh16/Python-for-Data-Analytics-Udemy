### .at[ ] and .iat[ ]
the above operation can be simplified
- .at - using the labels as locations, even integers are treated as labels
- .iat - using the integer based locations

using **iat[ ]**

```python
stats.iat[1,3]
```

using **at[ ]**

```python
stats.at[1,'InternetUsers']
```

**let's see the difference between .at[ ] and .iat[ ]**

```python
stats_2 = stats[::10]
stats_2
```

```python
stats_2.iat[10, 0]
```

```python
stats_2.at[10, 'CountryName']
```

why the results are different? it is because the **.iat[ ]** potentially uses the integer-based traverse and **.at[ ]** uses the labels. though the integer 10 is used in .at[], it is used as a label and not as integer index and returns the corresponding data.

### using .loc[ ] and .iloc[ ]
a series of rows and columns can be accessed by:
- .loc - using the labels as locations, even integers are treated as labels
- .iloc - using the integer based locations

using **loc[ ]**

```python
#Accessing a set of rows from the dataset
stats.loc[0:10]
```

```python
#Accessing a single column from the dataset
stats.loc[:, 'CountryName']
```

```python
#Accessing a set of columns from the dataset
stats.loc[:, ['CountryName', 'CountryCode']]
```

```python
#Accessing a subset of the dataset
stats.loc[0:10, ['CountryName', 'CountryCode']]
```

using **iloc[ ]**

```python
#Accessing a set of rows from the dataset
stats.iloc[0:10]
```

```python
#Accessing a single column from the dataset
stats.iloc[:, 0]
```

```python
#Accessing a set of columns from the dataset
stats.iloc[:, [0, 1]]
```

```python
#Accessing a subset of the dataset
stats.iloc[0:10, [0, 1]]
```

**let's see the difference between .loc[ ] and .iloc[ ]**

```python
stats_3 = stats[::10]
stats_3
```

```python
stats_3.loc[10, 'CountryName']
```

```python
stats_3.iloc[10, 0]
```

why the results are different? it is because the **.iloc[ ]** potentially uses the integer-based traverse and **.loc[ ]** uses the labels. though the integer 10 is used in .loc[ ], it is used as a label and not as integer index and returns the corresponding data.
