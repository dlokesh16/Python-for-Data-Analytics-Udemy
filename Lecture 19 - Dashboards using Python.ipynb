```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 4
warnings.filterwarnings('ignore')
```

```python
movies = pd.read_csv('Data/Movie-Ratings-Data.csv')
```

```python
movies.columns = ['Film', 'Genre', 'CriticRating', 'AudienceRating', 'BudgetMillions', 'Year']
```

```python
movies.Genre = movies.Genre.astype('category')
movies.Year = movies.Year.astype('category')
```

### Dashboards
When we have a visualization with multiple kinds of plots, then we can consider that as a dashboard.

```python
f, axes = plt.subplots(2, 2, figsize = (12, 12))

a1 = sns.kdeplot(data = movies, x = 'BudgetMillions', y = 'AudienceRating', ax = axes[0, 0])
a2 = sns.kdeplot(data = movies, x = 'BudgetMillions', y = 'CriticRating', ax = axes[0, 1])

a3 = sns.violinplot(data = movies[movies.Genre == 'Drama'], x = 'Year', y = 'CriticRating', ax = axes[1, 0])

a4 = sns.kdeplot(data = movies, x = 'CriticRating', y = 'AudienceRating', fill = True, cmap = 'Reds', shade_lowest = False, ax = axes[1,1])
a4s = sns.kdeplot(data = movies, x = 'CriticRating', y = 'AudienceRating', cmap = 'Reds', ax = axes[1,1])

a1.set_xlim(-40, 260)
a2.set_xlim(-40, 260)

plt.show()
```

Seaborn and Pyplot uses different approaches to plot to the subplots.<br>
Seaborn plot accept an input **axis** parameter to plot on the respective subplot.<br>
Pyplot uses the **axis object** to plot on the respective subplot (as mentioned below).

```python
f, axes = plt.subplots(2, 2, figsize = (12, 12))

a1 = sns.kdeplot(data = movies, x = 'BudgetMillions', y = 'AudienceRating', ax = axes[0, 0])
a2 = sns.kdeplot(data = movies, x = 'BudgetMillions', y = 'CriticRating', ax = axes[0, 1])

#a3 = sns.violinplot(data = movies[movies.Genre == 'Drama'], x = 'Year', y = 'CriticRating', ax = axes[1, 0])
axes[1, 0].hist(movies.CriticRating, bins = 15)

a4 = sns.kdeplot(data = movies, x = 'CriticRating', y = 'AudienceRating', fill = True, cmap = 'Reds', shade_lowest = False, ax = axes[1,1])
a4s = sns.kdeplot(data = movies, x = 'CriticRating', y = 'AudienceRating', cmap = 'Reds', ax = axes[1,1])

a1.set_xlim(-40, 260)
a2.set_xlim(-40, 260)

plt.show()
```
