```python
#Function takes the Matrix to visualize and the PlayerList as inputs. By default, Points matrix will be visualized and Players list is passed
def customplot(InputMatrix = Points, PlayerList = Players):
    ColorList = { "KobeBryant":"Black","JoeJohnson":"Red","LeBronJames":"Green","CarmeloAnthony":"Blue","DwightHoward":"Magenta","ChrisBosh":"Black","ChrisPaul":"Red","KevinDurant":"Green","DerrickRose":"Blue","DwayneWade":"Magenta" }
    MarkerList = { "KobeBryant":"s","JoeJohnson":"o","LeBronJames":"^","CarmeloAnthony":"D","DwightHoward":"s","ChrisBosh":"o","ChrisPaul":"^","KevinDurant":"D","DerrickRose":"s","DwayneWade":"o" }
    for name in PlayerList:
        plt.plot(InputMatrix[Pdict[name]], c=ColorList[name], ls='--', marker=MarkerList[name], ms=5, label=name)
    plt.xticks(list(range(0, 10)), Seasons)
    plt.legend(loc = 'upper left', ncol = 2, bbox_to_anchor=(1,1))
    plt.show()
```

Explanation for the above custom function
- **customplot()** function takes a Matrix to visualize as InputMatrix and the list of players as PlayerList, as inputs
- **ColorList** internal list is created to assign a particular color to each of the player, so that the graph lines are consistent
- **MarkerList** internal list is created to assign a particular marker style to each of the player, so that the graph markers are consistent
- Plots are created by iterating through the PlayerList input and for the particular InputMatrix
- Other plot aesthetics are performed

```python
#Free Throw Attempts per game
customplot(FreeThrowAttempts / Games)
```

From the above plot, it is clear that Dwight Howard has the most number of Free Throw attempts and Joe Johnson has the least number of Free Throw attempts

```python
#Accuracy of Free Throws
customplot(FreeThrows / FreeThrowAttempts)
```

From the above plot, it is clear that Dwight Howard has the least accuracy of Free Throws and Kevin Durant has one of the best Free Throws

```python
#Player playing style (2 vs 3 points preference) excluding Free Throws (Each Free Throw is worth 1 point)
customplot((Points - FreeThrows) / FieldGoals)
```

Because we have excluded free throws, the above plot now shows us the true representation of player style change. We can verify
that this is the case because all the marks without exception on this plot are between 2 and 3. That is because Field Goals can only be for either 2 points or 3 points.

Insights:
1. You can see how players' preference for 2 or 3 point shots changes throughout their career. We can see that almost all players in this dataset experiment with their style throughout their careers. Perhaps, the most drastic change in style has been experienced by Joe Johnson.

2. There is one exception. You can see that one player has not changed his style at all - almost always scoring only 2-pointers. It's Dwight Howard! Now that explains a lot. The reason that Dwight Howard's Field Goal accuracy is so good is because he almost always scores 2-pointers only. That means he can be close to the basket or even in contact with it. Free throws, on the other hand require the player to stand 15ft (4.57m) away from the hoop. That's probably why Dwight Howard's Free Throw Accuracy is poor.
