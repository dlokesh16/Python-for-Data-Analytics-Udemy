```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 4
warnings.filterwarnings('ignore')
```

```python
movies = pd.read_csv('Data/Movie-Ratings-Data.csv')
```

```python
movies.columns = ['Film', 'Genre', 'CriticRating', 'AudienceRating', 'BudgetMillions', 'Year']
```

```python
movies.Genre = movies.Genre.astype('category')
movies.Year = movies.Year.astype('category')
```

### Subplots
To create subplots, we need to make use of matplotlib

```python
k1 = sns.kdeplot(data = movies, x = 'BudgetMillions', y = 'AudienceRating')
```

```python
k2 = sns.kdeplot(data = movies, x = 'BudgetMillions', y = 'CriticRating')
```

```python
f, axes = plt.subplots(1, 2, figsize = (12, 6), sharex = True, sharey = True)
k1 = sns.kdeplot(data = movies, x = 'BudgetMillions', y = 'AudienceRating', ax = axes[0])
k2 = sns.kdeplot(data = movies, x = 'BudgetMillions', y = 'CriticRating', ax = axes[1])
k1.set_xlim(-40, 260)
plt.show()
```

From the above plot, we can clearly see that Audience ratings are much concentrated around 40-60 and Critic ratings are much concentrated around 20-60. Also, in both cases, the lowest ratings tend to increase and highest ratings tend to decrease post 40 Million budget. Also, it can be inferred higher buget movies does not always result in high ratings.
