### Normal Distribution
Also known as **Guassian Distribution**, is a distribution in which the **data is distributed equally on both sides of 0**. Also, **68.2%** of the data falls within **1 SD** (34.1% on either side of 0), **95.4%** of the data falls within **2 SD** (47.7% on either side of 0) and **99.6%** of the data falls within **3 SD** (49.8% on either side of 0)



---

In this exercise, we are going to perform the below steps to validate the **LLN** over a **Normal Distribution**



---

```python
import numpy as np
from numpy.random import randn as rand
```

```python
N = 100000                                                                       #initialize N, total sample count
counter = 0                                                                      #initialize counter, count of data falling between -1 and 1
for i in rand(N):                                                                #for each randomly generated number
    if (i >= -1 and i <=1):                                                      #check if data falls between -1 and 1
        counter += 1                                                             #increase counter
print('Count of Data within range -1 and 1 is : ', counter)                      #print count of data within range -1 and 1 
print('% of Data within range -1 and 1 is : ', round(counter / N * 100, 2), '%') #print % of data within range -1 and 1
```
