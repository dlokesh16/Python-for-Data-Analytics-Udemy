```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 4
warnings.filterwarnings('ignore')
```

```python
movies = pd.read_csv('Data/Movie-Ratings-Data.csv')
```

```python
movies.columns = ['Film', 'Genre', 'CriticRating', 'AudienceRating', 'BudgetMillions', 'Year']
```

```python
movies.Genre = movies.Genre.astype('category')
movies.Year = movies.Year.astype('category')
```

### FacetGrid

```python
viz1 = sns.FacetGrid(data = movies, row = 'Genre', col = 'Year')
viz1 = viz1.map(sns.scatterplot, 'CriticRating', 'AudienceRating')
```

The above facet grid gives a very detailed insights about the critic ratings, audience ratings for each genre and each year.

```python
viz2 = sns.FacetGrid(data = movies, row = 'Genre', col = 'Year')
viz2 = viz2.map(sns.histplot, 'BudgetMillions')
```

```python
#Let's do some tidying up on the FacetGrid with scatterplots
viz3 = sns.FacetGrid(data = movies, row = 'Genre', col = 'Year')
kws = dict(s = 50, linewidth = 0.5, edgecolor = 'black') #keyword argument to increase the dot size, and give an outline colour
viz3 = viz3.map(sns.scatterplot, 'CriticRating', 'AudienceRating', **kws)
viz3.set(xlim = (0, 100), ylim = (0, 100))

for ax in viz3.axes.flat:
    ax.plot((0, 100), (0, 100), c = 'black', ls = '--')
    
viz3.add_legend()
```

The above final plot give a much more clear insights. The diagonal gives us an idea on the CriticRating vs AudienceRating. Which group rated higher or lower for respective genres and respective years.
<br>For eg: 'Action' genre in 2009, Audience have given more lower ratings that Critics.
<br>But for 'Thriller' genre in 2011, Critics have given more high ratings that Audience, and that too really high around 80s and 90s.
